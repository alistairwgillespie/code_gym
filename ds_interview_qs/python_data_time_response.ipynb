# Time for a response on a messaging application

You are given a dataset
(https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/sample_message_dataset.csv
with information around messages sent between users in a P2P messaging application. Below is the dataset's schema:

Given this, write code to find the fraction of messages that are sent between the same sender and receiver within five
minutes (e.g. the fraction of messages that receive a response within 5 minutes).

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/sample_message_dataset.csv'
```

```python
df = pd.read_csv(url)
```

```python
display(df)
```

```python
message_lookup = {}
```

```python
df = df.sort_values(by=['date', 'timestamp'])
```

```python
df.shape
```

```python
df = df[df.sender_id != df.receiver_id]
```

```python
df.shape
```

```python
df.shape
```

```python
count = 0
for i,x in df.iterrows():
    message_lookup[x['sender_id']] = [x['receiver_id'], x['timestamp']]
    if x['receiver_id'] in message_lookup and x['sender_id'] == message_lookup[x['receiver_id']][0]:
        time_el = x['timestamp'] - message_lookup[x['receiver_id']][1]
        if time_el <= 300:
            count +=1
    message_lookup.pop(x['receiver_id'], None)
```

```python
count / 86 * 100
```

```python
df['sender_id_lag'] = df['sender_id'].shift(1)
```
